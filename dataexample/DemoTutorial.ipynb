# HOW TO CREATE A DATAFRAME....TWO DIMENSIONAL ARRAY

```python
# creating a dataframe using a list
data = [1,2,3,4]
df =  pd.DataFrame(data)
df
```

```python
# creating a dataframe using a dictionary
dictionary = {'fruits' : ['apples', 'banana', 'mangoes'], 'count' : [60,50,45]}
df = pd.DataFrame(dictionary)
df
```

```python
# creating a dataframe using series
series = pd.Series([6,12], index = ['a','b'])
df = pd.DataFrame(series)
df
```

```python
# creating a dataframe using a numpy array
import numpy as np
numpyarray = np.array([[50000,60000], ['John','James']])
df = pd.DataFrame({'name':numpyarray[1], 'salary':numpyarray[0]})
df
```

# HOW TO PERFORM A MERGE OPERATION

```python
player = ['player1','player2','player3']
point = [7,8,9,]
title = ['Game1', 'Game2', 'Game3']
df1 = pd.DataFrame ({'Player':player, 'Point':point, 'Title':title})
df1
```

```python
player = ['player1','player2','player3']
power = ['punch','kick','elbow',]
title = ['Game1', 'Game5', 'Game6']
df2 = pd.DataFrame ({'Player':player, 'Power':power, 'Title':title})
df2
```

```python
# inner merge
df1.merge(df2, on = 'Title', how ='inner')
```

```python
# left merger
df1.merge(df2, on = 'Title', how ='left')
```

```python
# Right merger
df1.merge(df2, on = 'Title', how ='right')
```

# HOW TO PERFORM JOIN OPERATION IN PANDAS

```python
player = ['player1','player2','player3']
point = [8,9,6,]
title = ['Game1', 'Game2', 'Game3']
df3 = pd.DataFrame ({'Player':player, 'Points':point, 'Title':title}, index = ['l1','l2','l3'])
df3
```

```python
player = ['player1','player5','player6']
power = ['punch','kick','elbow',]
title = ['Game1', 'Game5', 'Game6']
df4 = pd.DataFrame ({'Players':player, 'Power':power, 'Titles':title}, index = ['l2','l3','l4'])
df4
```

```python
df3.join(df4, how = 'inner')
```

```python
df3.join(df4, how = 'left')
```

# HOW TO CONCARTENAE TWO DFS IN PANDAS

```python
pd.concat([df3,df4])
```
