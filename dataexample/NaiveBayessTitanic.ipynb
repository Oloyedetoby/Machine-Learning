```python
import pandas as pd
```

```python
df = pd.read_csv('TitanicTrain.csv')
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
Scaler.fit(df[['Fare']])
df[['Fare']] = Scaler.transform(df[['Fare']])

Scaler.fit(df[['Survived']])
df[['Survived']] = Scaler.transform(df[['Survived']])
df
```

```python
from matplotlib import pyplot as plt
%matplotlib Inline
plt.scatter(df.Survived,df.Fare)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(df.Survived,df.Fare)
```

```python
df.drop(['Name', 'SibSp','Parch','Ticket','Embarked', 'PassengerId'], axis = 'columns' , inplace = True)
```

```python
df
```

```python
target = df.Survived
inputs = df.drop (['Survived','Cabin'],axis = 'columns')
```

```python
df.Cabin
```

```python
inputs
```

```python
dummies = pd.get_dummies(inputs.Sex)
```

```python
dummies
```

```python
inputs1 = pd.concat([inputs,dummies],axis = 'columns')
```

```python
inputs2 = inputs1.drop(['Sex'], axis = 'columns')
```

```python
inputs2
```

```python
target
```

```python
inputs2.columns[inputs2.isna().any()]
```

```python
inputs2.Age = inputs.Age.fillna(inputs2.Age.mean())
```

```python
inputs2
```

```python
inputs2.columns[inputs2.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs2,target, test_size = 0.3)
```

```python
len(inputs2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x_test[:10]
```

```python
y_test[:10]
```

```python
model.predict(x_test[:10])
```

```python
model.predict_proba(x_test[:10])
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cr = confusion_matrix(y_test, y_predicted)
cr
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cr, annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```
